```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def reduce_contrast(image, factor=0.5):
    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    min_val = np.min(image)
    max_val = np.max(image)

    reduced_image = ((image - min_val) * factor) + min_val + ((1 - factor) * max_val)
    reduced_image = np.clip(reduced_image, 0, 255).astype(np.uint8)

    return reduced_image

def histogram_equalization(image):
    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    equalized_image = cv2.equalizeHist(image)

    hist, _ = np.histogram(image.flatten(), 256, [0, 256])

    # Calculate cumulative distribution function (CDF)
    cdf = hist.cumsum()

    cdf_normalized = cdf / float(cdf.max())

    return equalized_image, cdf_normalized

def display_with_histogram_and_intensity_transformation(image, cdf_normalized):
    hist = cv2.calcHist([image], [0], None, [256], [0, 256])

    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title('Image')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.plot(hist, color='black')
    plt.title('Histogram')
    plt.xlabel('Pixel intensity')
    plt.ylabel('Frequency')

    plt.subplot(1, 3, 3)
    plt.plot(cdf_normalized, color='blue')
    plt.title('Intensity Transformation')
    plt.xlabel('Input intensity (u)')
    plt.ylabel('Output intensity (v)')

    plt.show()

image1 = cv2.imread('image1.jpg')
image2 = cv2.imread('image2.jpg')
image3 = cv2.imread('image3.jpg')

low_contrast_image1 = reduce_contrast(image1)
low_contrast_image2 = reduce_contrast(image2)
low_contrast_image3 = reduce_contrast(image3)

equalized_image1, cdf_normalized1 = histogram_equalization(low_contrast_image1)
equalized_image2, cdf_normalized2 = histogram_equalization(low_contrast_image2)
equalized_image3, cdf_normalized3 = histogram_equalization(low_contrast_image3)

display_with_histogram_and_intensity_transformation(image1, cdf_normalized1)
display_with_histogram_and_intensity_transformation(image2, cdf_normalized2)
display_with_histogram_and_intensity_transformation(image3, cdf_normalized3)
```
