```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

image1_path = 'image1.jpg'
image2_path = 'image2.jpg'

image1 = cv2.imread(image1_path, cv2.IMREAD_COLOR)
image2 = cv2.imread(image2_path, cv2.IMREAD_COLOR)

if image1.shape[0] > image2.shape[0] or image1.shape[1] > image2.shape[1]:
    image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))
else:
    image1 = cv2.resize(image1, (image2.shape[1], image2.shape[0]))

image1_float32 = image1.astype(np.float32)
image2_float32 = image2.astype(np.float32)

image1_log = np.log(image1_float32 + 1)
image2_log = np.log(image2_float32 + 1)

addition_result = np.log(image1_float32 + 255) + np.log(image2_float32 + 255)
subtraction_result = np.log(np.abs(image1_float32 - image2_float32) + 150)
scalar = 1.5
scalar_multiplication_result = np.log(image1_float32 * scalar + 1)
multiplication_result = np.log(image1_float32 * image2_float32 + 254)

darken_factor = 0.5
dark_image1 = np.clip(image1_float32 * darken_factor, 0, 255).astype(np.uint8)
dark_image2 = np.clip(image2_float32 * darken_factor, 0, 255).astype(np.uint8)

lighten_factor = 1.5
light_image1 = np.clip(image1_float32 * lighten_factor, 0, 255).astype(np.uint8)
light_image2 = np.clip(image2_float32 * lighten_factor, 0, 255).astype(np.uint8)

gray_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
gray_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

cv2_imshow(dark_image1)
cv2_imshow(dark_image2)
cv2_imshow(light_image1)
cv2_imshow(light_image2)
cv2_imshow(gray_image1)
cv2_imshow(gray_image2)

cv2_imshow(np.exp(addition_result).astype(np.uint8))
cv2_imshow(np.exp(subtraction_result).astype(np.uint8))
cv2_imshow(np.exp(scalar_multiplication_result).astype(np.uint8))
cv2_imshow(np.exp(multiplication_result).astype(np.uint8))

```
